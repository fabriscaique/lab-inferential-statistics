## Lab | Inferential Statistics

---

It is assumed that the mean systolic blood pressure is Î¼ = 120 mm Hg. In the Honolulu Heart Study, a sample of n = 100 people had an average systolic blood pressure of 130.1 mm Hg with a standard deviation of 21.21 mm Hg. 

- Is the group significantly different (with respect to systolic blood pressure!) from the regular population?

Set up the hypothesis test.
Write down all the steps followed for setting up the test.
Calculate the test statistic by hand and also code it in Python. It should be 4.76190. We will take a look at how to make decisions based on this calculated value.

**1 | Hypothesis Statements**

- **($H_0$)**: $\mu_{\text{(global BP)}} = \mu_{\text{(Honolulu BP)}}$  
  *(There is no difference in means between Global Blood Pressure and Honolulu Blood Pressure.)*

- **($H_1$)**: $\mu_{\text{(global BP)}} \neq \mu_{\text{(Honolulu BP)}}$  
  *(There is a difference in means between Global BLood Pressure and Honolulu Blood Pressure.)*

**2 | Significance Level**

Will be set at $\alpha = 0.05$

**3 | Calculate the Test Statistics**

$$
t = \frac{\bar{x} - \mu}{\frac{\text{std}}{\sqrt{n}}}
$$

```python
# Data
mu = 120 #mm Hg (expected systolic blood pressure)
n = 100 #sample size
x = 130.1 #mm Hg (Honolulu study mean)
std = 21.21 #mm Hg (Standard Deviation)
a = 0.05
```

```python
t_stat = (x - mu) / (std / np.sqrt(n))

if t_stat > 4.7 < 4.8:
    print('Test statistic matches Lab expectations.')
    print(f'T-test is {round(t_stat, 5)}')
```

```python
crit_value = st.norm.ppf(a)
round(crit_value, 2)
```

```python
p_value = 1 - st.norm.cdf(t_stat)

if p_value < 0.5:
    print('Null Hypothesis Rejected')
```

```python
# Viz
x = np.linspace(-3, 3, 100)
y = st.norm.pdf(x)
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, 0, y, where=(x < crit_value), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(t_stat, color='blue', linestyle='--', label=f'Test statistic (t = {round(t_stat, 5)})')
plt.title('Hypothesis Testing Visualization', fontsize=14)
plt.legend(loc='lower center')
plt.show()
```
